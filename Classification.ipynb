#### Data Transformation

### <font color='blue'> binomial column mapping Sex has a value of 1 and a value of 0 </font> 

```python
traindata['Sex']=traindata['Sex'].map({'m':1 , 'f':0})
testdata['Sex']=testdata['Sex'].map({'m':1 , 'f':0})
```

#### Because it is blood donors vs. Hepatitis C, including its progression ('just' Hepatitis C, Fibrosis, Cirrhosis), blood donors and suspected blood donors have 1 and the rest have zero

```python
traindata['Category']=traindata['Category'].map({'0=Blood Donor':1,
'3=Cirrhosis':0,'1=Hepatitis':0,'2=Fibrosis':0,'0s=suspect Blood Donor':1})
```

```python
traindata.info()
```

```python
testdata.info()
```

#### Separation of the feature set and the target set, as well as the filling of NaN values with the supervised learning model KNN imputer

#### "PROT" and "ALB" dropping the row because they have 3 NaN in this column if everything is imputed by mean or median, which will dissort the data structure.

```python
traindata.dropna(subset=['PROT','ALB','ALT'],inplace=True)
```

```python
target=traindata['Category']
#### Drop the Category from the traindataset
traindata.drop('Category',axis=1,inplace=True)
```

```python
### importing imputer library
from sklearn.impute import KNNImputer
### creating Object with default properties
imputer = KNNImputer()
#### fitting according to traindata and Transforming it 
traindata=pd.DataFrame(imputer.fit_transform(traindata),columns=traindata.columns)
### transforming test data according to the train data 
testdata=pd.DataFrame(imputer.transform(testdata),columns=testdata.columns)
```

```python
### Importing Split libraries
from sklearn.model_selection import train_test_split
#split the dataset into training (80%) and testing (20%) sets
X_train,X_valid,y_train,y_valid = train_test_split(traindata,target,test_size=0.20,random_state=0,stratify=target) 
```

```python
### ploting Correlation 
%matplotlib inline
plt.imshow(np.corrcoef(X_train.T))
plt.xticks(np.arange(12),testdata.columns, rotation = 90)
plt.yticks(np.arange(12),testdata.columns)
plt.colorbar()
plt.show()
```

#### Applying the Logistic Regression Model, Since this is a recall of oriented machine learning task setting parametters according

```python
### importing logistic regression
from sklearn.linear_model import LogisticRegression

#creating model object
lreg = LogisticRegression(max_iter = 1000, random_state= 0)

#fit the model using the training data
lreg.fit(X_train,y_train)
```

```python
#use model to make predictions on validation dataset
y_pred = lreg.predict(X_valid)
```

```python
from sklearn.metrics import confusion_matrix

print('Train confusion matrix: ')
print(confusion_matrix(y_train, lreg.predict(X_train)))
print('Test confusion matrix: ')
print(confusion_matrix(y_valid, lreg.predict(X_valid)))

y_pred=lreg.predict(X_valid)
```

```python
from sklearn.metrics import accuracy_score,recall_score, precision_score, f1_score

print('Macro precision: ', precision_score(y_valid,y_pred, average= 'macro'))
print('Micro precision: ', precision_score(y_valid, y_pred, average= 'micro'))
print('Macro recall: ', recall_score(y_valid, y_pred, average= 'macro'))
print('Micro recall: ', precision_score(y_valid, y_pred, average= 'micro'))
print('Macro f1-score: ', f1_score(y_valid, y_pred, average= 'macro'))
print('Micro f1-score: ', f1_score(y_valid,y_pred, average= 'micro'))
```

```python
print("Accuracy on test data: {:.3f}".format(accuracy_score(y_valid,y_pred)))
print("Accuracy on training set: {:.3f}".format(accuracy_score(y_train,lreg.predict(X_train))))
```

```python
%matplotlib inline
confusion = confusion_matrix(y_valid, y_pred)

fig, ax = plt.subplots()
plt.imshow(confusion)
plt.title("Confusion matrix")

ax.set_xticks(np.arange(len(confusion)))
ax.set_yticks(np.arange(len(confusion[0])))

# Loop over data dimensions and create text annotations.
for i in range(len(confusion)):
    for j in range(len(confusion[0])):
        text = ax.text(j, i, confusion[i, j],
                       ha="center", va="center", color="w")
fig.tight_layout()
plt.xlabel('Actual')
plt.ylabel('Prediction')
plt.show()
```

#### Applying the decision tree classifier model, Since this is a recall of oriented machine learning task setting parametters

```python
### importing tree
from sklearn.tree import DecisionTreeClassifier
### object declaration
decision_tree = DecisionTreeClassifier(random_state=0,criterion= 'entropy', min_samples_split=5,max_depth=4)
### fitting train data
decision_tree.fit(X_train,y_train)
```

```python
from sklearn import tree 

plt.subplots(figsize = (15,15))
tree.plot_tree(decision_tree, 
               feature_names=X_train.columns,  
               filled=True,
               rounded = True)
plt.show()
```

```python
print('Train confusion matrix: ')
print(confusion_matrix(y_train, decision_tree.predict(X_train)))
print('Test confusion matrix: ')
print(confusion_matrix(y_valid, decision_tree.predict(X_valid)))

pred=decision_tree.predict(X_valid)
```

```python
print("Accuracy on training set: {:.3f}".format(decision_tree.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(decision_tree.score(X_valid, y_valid)))
```

```python
print('Macro precision: ', precision_score(y_valid,pred, average= 'macro'))
print('Micro precision: ', precision_score(y_valid, pred, average= 'micro'))
print('Macro recall: ', recall_score(y_valid, y_pred, average= 'macro'))
print('Micro recall: ', precision_score(y_valid, pred, average= 'micro'))
print('Macro f1-score: ', f1_score(y_valid, pred, average= 'macro'))
print('Micro f1-score: ', f1_score(y_valid,pred, average= 'micro'))
```
